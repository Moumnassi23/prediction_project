```python
import pandas as pd
import numpy as np
from pycaret.datasets import get_data
from pycaret.classification import * 
```

```python
get_data("index")
```

```python
from sklearn.model_selection import train_test_split
data = get_data("hepatitis", verbose=False)
data, data_unseen = train_test_split(data, test_size=0.2, random_state=123)
data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)
data.shape, data_unseen.shape


```

```python
exp_1 = ClassificationExperiment()

exp_1.setup(data=data, target='Class', session_id=123, verbose=True)
best = exp_1.compare_models()
```

```python
#EDA with pycaret
from ydata_profiling import ProfileReport
profile = ProfileReport(data, title="Hepatitis Dataset Profiling Report", explorative=True)
profile.to_file("rapport.html")

```
